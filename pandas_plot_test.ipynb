```python
import pandas as pd
import seaborn as sns
from mpl_toolkits.mplot3d import Axes3D

df = pd.read_csv('mtcars.csv')
sns.set(style = "darkgrid")

fig = plt.figure()
ax = fig.add_subplot(111, projection = '3d')

x = df['cyl']
y = df['hp']
z = df['wt']

ax.set_xlabel("cyl")
ax.set_ylabel("hp")
ax.set_zlabel("wt")

ax.scatter(x, y, z)

plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

df = pd.read_csv('mtcars.csv')

grid = sns.FacetGrid(df, col = "model", hue = "model", col_wrap=5)
grid.map(sns.scatterplot, "hp", "wt")

grid.add_legend()

plt.show()
```
